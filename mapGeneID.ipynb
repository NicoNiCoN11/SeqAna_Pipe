```python
import mygene
import pandas as pd
```

```python
gene_list = pd.read_csv('/home/jiguo/SeqAna_Pipe/significant_genes_RPE-p53KO_vs_Control.txt', sep='\t', index_col=0)
mg = mygene.MyGeneInfo()
gene_ids = [g.split(".")[0] for g in gene_list.index]
```

```python

gene_ids = gene_list.index.str.split('.').str[0].tolist()
res = mg.querymany(
    gene_ids,
    scopes="ensembl.gene",
    fields="symbol,name",
    species="human"
)

annotations = pd.DataFrame(res)[
    ["query", "symbol", "name"]
].rename(
    columns={"query": "EnsemblID", "symbol": "GeneSymbol", "name": "Description"}
)
```

```python
gene_list["EnsemblID"] = gene_list.index.str.split('.').str[0]

final_df = gene_list.merge(
    annotations,
    on="EnsemblID",
    how="left"
)
final_df.to_csv('./significant_genes_RPE-p53KO_vs_Control_annotated.txt', sep='\t', index=True)
```
